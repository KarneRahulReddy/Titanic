```python
def age(titanic):
    Age = titanic[0]
    Pclass = titanic[1]
    Sex=titanic[2]
    if pd.isnull(Age):

        if ((Pclass == 1)&(Sex=='male')):
            return 40
        
        elif ((Pclass == 1)&(Sex=='female')):
            return 35

        elif ((Pclass == 2)&(Sex=='male')):
            
            return 30
        elif ((Pclass == 2)&(Sex=='female')):
            return 28

        elif ((Pclass == 3)&(Sex=='male')):
           
            return 25
        elif ((Pclass == 3)&(Sex=='female')):
            return 22

    else:
        return Age
```

```python
train['Age'] = train[['Age','Pclass','Sex']].apply(age,axis=1)
```
